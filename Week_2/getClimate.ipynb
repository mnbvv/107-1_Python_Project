# getClimate
---
### getClimate(position, date, info)
* postition : City  (ex. 台北市)
* date : YYYY-MM-DD (ex. 2018-01-01)
* info : station_info.csv

```python
import pandas as pd
import numpy as np
```

```python
# Set values for example
position = '台北市'
date = '2018-01-01'
info = pd.read_csv('station_info.csv')
info
```

```python
precipitation = []
temperature = []
```

```python
if position[0] == '台':
    position = '臺' + position[1:]
print(position)
```

for i in info[info['City']==position].index:  
           
        data = pd.read_html(info.loc[i,'Url'] + date)[1].iloc[2:,[3,10]] # 3:temp, 10:precip
            
        temp = pd.to_numeric(data[3], errors = 'coerce', downcast = 'float')
        if temp.count()!=0:
            temperature.append(temp.mean())
        temp = pd.to_numeric(data[10], errors = 'coerce', downcast = 'float')
        if temp.count()!=0:
            precipitation.append(temp.mean())
